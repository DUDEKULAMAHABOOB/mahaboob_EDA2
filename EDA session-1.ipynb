```python
Names=['Ramesh','Suresh','Sathish']
Age=[20,25,30]
df1=pd.DataFrame(zip(Names,Age),
             columns=['Names','Age'],
             index=[0,1,2])
df1['City']=['Hyd','Chennai','Blr']
df1
```
